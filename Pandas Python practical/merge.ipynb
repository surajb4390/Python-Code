```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'id':[1,2,3,4],'Name':['A','B','C','D']})
df1
```

```python
df2 = pd.DataFrame({'id':[1,2,3,4],'Class':[9,10,11,12]})
df2
```

```python
pd.merge(df1,df2,on = 'id')
```

```python
pd.merge(df2,df1, on = 'id') #here change position of parameter - df1,and df2 
```

```python
df3 = pd.DataFrame({'id':[1,2,3,5],'Class':[9,10,11,12]})
```

```python
df3
```

```python
pd.merge(df1,df3,on = 'id') #in df3 we change id value 4 = 5 then we get result only same no of id exicute here
```

```python
pd.merge(df1,df3,on = 'id',how  = 'inner')
```

```python
pd.merge(df1,df3,on = 'id',how = 'left')
```

```python
pd.merge(df1,df3,on = 'id',how = 'right')
```

```python
pd.merge(df1,df3,on = 'id',how = 'outer')
```

```python
pd.merge(df1,df3,on = 'id',how = 'outer',indicator=True)
```
